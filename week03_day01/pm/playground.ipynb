```python
movie = pd.read_csv('https://github.com/MahnoorJaved98/Movie-Recommendation-System/blob/main/movie_dataset.csv?raw=true').dropna()
movie
```

```python
genres = ''
for row in movie['genres']:
    genres += row + ' '

genres = list(set(genres.split(' ')[:-1]))
genres
```

```python
#   this movie is not an Action movie
# [ |
#  [0,1], -> Action
#  [1,0]  -> Adventure
#   |
#   this movie is an Adventure movie
# ]
genres_movies = [[] for i in range(len(genres))]

for row in movie['genres']:
    for index, value in enumerate(genres):
        if value in row.split(' '):
            genres_movies[index].append(1)
        else:
            genres_movies[index].append(0)
```

```python
genres_movies_dataframe = pd.DataFrame(np.array(genres_movies).T, columns=genres)
genres_movies_dataframe
```

```python
temp_dataframe = pd.DataFrame({'original_title': ['Avatar', np.nan, 'Avatar']}).dropna()
temp_dataframe['original_title'].reset_index(drop=True)
```

```python
title_df = movie['original_title'].reset_index(drop=True)
title_df
```

```python
movie_vector = pd.concat([title_df, genres_movies_dataframe], axis=1)
movie_vector.set_index('original_title', inplace=True)
movie_vector
```

```python
movie_vector.loc['Avatar'].to_list()
```

```python
def cosine_similarity(vect1, vect2):
    norm_1 = np.linalg.norm(vect1)
    norm_2 = np.linalg.norm(vect2)

    return (vect1 @ vect2) / (norm_1 * norm_2)
```

```python
movie.loc[movie['original_title'] == 'Avatar', ['original_title', 'genres']]
```

```python
movie.loc[movie['original_title'] == 'Tin Can Man', ['original_title', 'genres']]
```

```python
cosine_similarity(movie_vector.loc['Avatar'], movie_vector.loc['Tin Can Man'])
```

```python
def recommendation_system(movie, top_N):
    cos_values = pd.Series([cosine_similarity(movie_vector.loc[movie], x) for x in movie_vector.values], index=movie_vector.index) \
                   .drop(index=movie)
                   # drop if index = 'Man of Steel'
    print(cos_values)
    print(f'You like {movie}, so based on our recommender system, We recommend you to watch:')
    for index, value in enumerate(cos_values.sort_values(ascending=False)[:top_N].items()):
        print(f'{index+1}. {value[0]} {value[1]}')

recommendation_system('Man of Steel', 5)
```
