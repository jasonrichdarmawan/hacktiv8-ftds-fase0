```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/DianCA26/Dataset/main/CarPrice_Assignment.csv')
```

```python
df.columns
```

```python
df['CarName'].value_counts()
```

```python
plt.subplots(figsize=(20,5))
df['CarName'].value_counts().plot(kind='bar')
```

```python
df['carbody'].value_counts()
```

```python
plt.subplots(figsize=(20,5))
df['carbody'].value_counts().plot(kind='bar')
```

```python
import seaborn as sns
```

```python
var1=df['carbody'].value_counts()
var1.index, var1
```

```python
sns.barplot(y=var1.index,x=var1)
```

```python
b = df.groupby(['carbody','fueltype']).count()[['fuelsystem']].reset_index()
b
```

```python
sns.barplot(y=b['carbody'], x=b['fuelsystem'])
```

```python
df['carbody'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

```python
colors=sns.color_palette('bright')[5:10]
c = df['carbody'].value_counts()
plt.pie(c, colors=colors, autopct='%.2f%%')
plt.legend(c.index)
plt.show()
```

```python
df['price'].plot(kind='line')
```

```python
sns.lineplot(x=df['price'].index,y=df['price'])
```

```python
# correlation-nya rendah. terlihat dibawah, sangat menyebar.
display(df[['price', 'stroke']].corr())
fig, axes = plt.subplots(figsize=(10,5))
axes.scatter(x=df['price'],y=df['stroke'])
```

```python
display(df[['price', 'enginesize']].corr())
fig, axes = plt.subplots(figsize=(10,5))
axes.scatter(x=df['price'],y=df['enginesize'])
```

```python
fig, axes = plt.subplots(figsize=(10,5))
sns.scatterplot(x=df['price'],y=df['enginesize'], style=df['fueltype'])
```

```python
fig, axes = plt.subplots(figsize=(10,5))
sns.scatterplot(x=df['price'],y=df['enginesize'], hue=df['fueltype'])
```
