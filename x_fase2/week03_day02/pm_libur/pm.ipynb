```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    data={
        'fruit': [
            'Apple A', 'Apple B', 'Apple C', 'Apple D', 'Apple E', 'Apple F',
            'Banana', 'Carambola', 'Kiwi B', 'Kiwi C', 'Mango', 'Orange', 'Peach',
            'Pear', 'Persimmon', 'Pitaya', 'Plum', 'Pomegranate', 'Tomatoes',
            'guava A', 'guava B', 'kiwi A', 'muskmelon'
        ],
        'precision': [
            0.88, 0.84, 0.88, 0.93, 0.88, 0.63,
            0.71, 1, 0.89, 0.97, 0.87, 0.97, 0.95,
            0.79, 1, 0.91, 1, 0.91, 0.94,
            0.91, 0.88, 0.88, 0.90
        ],
        'recall': [
            0.88, 1, 0.91, 0.84, 0.91, 0.84,
            0.69, 1, 0.97, 0.97, 0.81, 1, 0.62,
            0.81, 0.97, 0.97, 1, 0.94, 0.94,
            0.97, 0.88, 0.91, 0.56
        ],
        'f1-score': [
            0.88, 0.91, 0.89, 0.89, 0.89, 0.72,
            0.7, 1, 0.93, 0.97, 0.84, 0.98, 0.75,
            0.8, 0.98, 0.94, 1, 0.92, 0.94,
            0.94, 0.88, 0.89, 0.69
        ]
    }
)

df['equal'] = df['f1-score'] == np.round(2 * df['precision'] * df['recall'] 
                                         / (df['precision'] + df['recall']), 2)

df.sort_values(by='f1-score', ascending=False) \
  .style.background_gradient(cmap='RdYlGn') \
  .format({'precision': "{:.2f}", 'recall': "{:.2f}", 'f1-score': "{:.2f}"})
```
