```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service as EdgeService
# EdgeChromiumDriverManager.install() is very slow in Mac Mini M1
# from webdriver_manager.microsoft import EdgeChromiumDriverManager

import pandas as pd

import selenium.common.exceptions as exc
```

```python
driver = webdriver.Edge(service=EdgeService(executable_path="/Volumes/T7Touch/Learn/hacktiv8-ftds-fase0/week03_day05/am/msedgedriver"))
```

```python
driver.get("https://www.imdb.com/search/title/?genres=animation")
```

```python
def webScraping():
    # reference: https://www.lambdatest.com/blog/complete-guide-for-using-xpath-in-selenium-with-examples/
    # XPath = //tagname[@Attribute='Value']

    # reference: https://www.javatpoint.com/xpath-expression
    # / -> It specifies that selection starts from the root node.
    # . -> Select the current node.
    # // -> It specifies that selection STARTS from the current node that match the selection.
    # @ -> Select attributes
    # p[...]/span -> selects all span elements that are children of p
    title = []
    rating = []
    description = []
    votes = []
    poster = []
    for index, lister_item in enumerate(driver.find_elements(by=By.XPATH, 
                                                             value="//div[@class='lister-item mode-advanced']")):
        title.append(lister_item.find_element(by=By.XPATH,
                                              value=".//h3[@class='lister-item-header']/a[1]")
                                .text)
        
        try:
            rating.append(lister_item.find_element(by=By.XPATH, 
                                        value=".//div[@class='inline-block ratings-imdb-rating'][1]")
                                     .get_attribute('data-value'))
        except exc.NoSuchElementException:
            rating.append(float('nan'))

        description.append(lister_item.find_element(by=By.XPATH,
                                                    value=".//p[@class='text-muted'][1]")
                                      .text)

        try:
            votes.append(lister_item.find_element(by=By.XPATH,
                                        value=".//span[@name='nv'][1]")
                                    .text)
        except exc.NoSuchElementException:
            votes.append(float('nan'))

        poster.append(lister_item.find_element(by=By.XPATH,
                                               value=".//img[@class='loadlate'][1]")
                                 .get_attribute("src"))

    display(pd.DataFrame({"Title": title, 
                          "Rating": rating, 
                          "Description": description, 
                          "Votes": votes, 
                          "Poster": poster}))

webScraping()
```
