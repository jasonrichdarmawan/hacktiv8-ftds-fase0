```python
import pandas as pd
import plotly.express as px
import folium
```

```python
df = pd.read_csv("../../week02_day03/h8dsft_Descriptive/covid_19_indonesia_time_series_all.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Location', 'Date'])
```

```python
series = df.loc[df['Location'].isin(['DKI Jakarta', 'Riau', 'Aceh']), 
                ['Location', 'New Cases']] \
           .groupby('Location') \
           .mean()
series
```

```python
px.bar(series, x=series.index, y='New Cases',
       labels={"New Cases": "New Cases Mean"})
```

```python
df3 = pd.read_csv("./bigquery-public-data.iowa_liquor_sales.sales.csv")
```

```python
df3.head()
```

```python
import re
```

```python
m2 = folium.Map(location=[42.224536, -92.868874], zoom_start=5)

for row in df3.dropna(subset=['store_location']) \
              .drop_duplicates(subset=['store_name'], keep='last') \
              .itertuples():
    try:
        longitude, langitude = re.findall("POINT\(([0-9-.]+) ([0-9-.]+)\)", 
                                          row.store_location)[0]
        folium.Marker(location=[langitude, longitude], 
                      popup=row.store_name, 
                      tooltip=row.store_name) \
              .add_to(m2)
    except TypeError:
        pass

m2
```
