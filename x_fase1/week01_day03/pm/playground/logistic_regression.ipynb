```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
X, y = load_iris(return_X_y=True)
```

```python
# 0 = iris sentoosa
# 1 = iris veronica
# 2 = iris virginica
y
```

```python
import numpy as np
import seaborn as sns

# sepal length
# sepal width
# petal length
# petal width
# target
#               sepal length | sepal width | petal length | petal width | target

columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'target']
corrcoef = np.corrcoef(X.T, y.T)
sns.heatmap(data=corrcoef, annot=True, xticklabels=columns, yticklabels=columns)
```

```python
X.shape
```

```python
corrcoef
```

```python
# sepal length = 0.78
# sepal width = -0.42
# petal length = 0.94 (high correlation to target)
# petal width = 0.95  (high correlation to target)
corrcoef[4]
```

```python
# get columns `petal length` and `petal width``
X[:,(2,3)]
```

```python
clf = LogisticRegression(random_state=0).fit(X[:,(2,3)], y)
```

```python
# get top 2 rows
X[:2,(2,3)]
```

```python
clf.predict(X[:2, (2,3)])
```

```python
# class 0 have probability > 0.5. Therefore:
# row 0 = iris sentosa
# row 1 = iris sentosa
clf.predict_proba(X[:2, (2,3)]) > 0.5
```

```python
clf.score(X[:,(2,3)], y)
```
