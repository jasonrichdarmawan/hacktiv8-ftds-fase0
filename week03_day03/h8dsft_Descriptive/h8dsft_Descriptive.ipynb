# Data Cleaning

## fix: column `Date`

```python
# cast column Date to datetime, so we can sort it later.
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df = df.sort_values(['Location ISO Code','Date'])
df.head()
```

```python
# drop unnecessary columns.
df2 = df.drop(df.columns[[1,11,12,13,14,15,16,17,18,27,28]],axis=1).copy()

# force pandas to display any number of columns
pd.set_option('display.max_columns', None)
df2
```

## fix: column `Total Cities`

```python
# row with column 'Location' == 'Sulawesi Barat' have only NaN value in column 'Total Cities'
# If we wish to find the central tendency of the dataset, 
# either drop the 'Total Cities' column or drop rows with column 'Location' == 'Sulawesi Barat'
# we will drop column 'Total Cities'
df2.loc[(df2['Total Cities'].isna()) & (df2['Location'] == 'Sulawesi Barat')].info()
df3 = df2.drop('Total Cities', axis='columns').copy()
df3
```

## fix: column `Total Urban Villages`

```python
# quick verdict: drop the column
df3.loc[(df3['Location'] == 'Aceh')].info()
df4 = df3.drop('Total Urban Villages', axis='columns').copy()
df4
```

## fix: column `Total Rural Villages`

```python
# quick verdict: drop the column
df4.loc[(df4['Location'] == 'DKI Jakarta')].info()
df5 = df4.drop('Total Rural Villages', axis='columns').copy()
df5
```

## fix: column `Growth Factor of New Cases`

```python
df5.loc[df5['Growth Factor of New Cases'].isna()] # Aceh, Daerah Istimewa Yogyakarta, Indonesia, ...

# NaN value because no previous data.
temp_df5_aceh = df5.loc[(df5['Location'] == 'Aceh'), 
        ['New Cases', 'Growth Factor of New Cases']]

# NaN value because Yesterday New Cases = 0
temp_df5_aceh.loc[df5['Growth Factor of New Cases'].isna()] # 238 988 1198 1408 1513
temp_df5_aceh.index.get_loc(988) # 23
df5.iloc[21:].loc[:,['New Cases', 'Growth Factor of New Cases']]

# verdict: replace NaN value with 0.
df5['Growth Factor of New Cases'] = df5['Growth Factor of New Cases'].fillna(0)
df5.info()
```

## fix: column `Growth Factor of New Deaths`

```python
## quick verdict: replace NaN value with 0

df5['Growth Factor of New Deaths'] = df5['Growth Factor of New Deaths'].fillna(0)
df5.info()
```
